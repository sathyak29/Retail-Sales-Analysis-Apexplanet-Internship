# STEP 2:-SQL for Business Questions

```python
import sqlite3

# Create in memory SQLite database
conn = sqlite3.connect(':memory:')
df.to_sql('Superstore_Sales',conn,index=False,if_exists='replace')
print("Database ready! Table: Superstore_Sales")
```

```python
# Q1: What are the Top 5 products by total revenue?
query1 = """
SELECT "Sub-Category", ROUND(SUM(Sales), 2) AS Total_Revenue
FROM superstore_Sales
GROUP BY "Sub-Category"
ORDER BY Total_Revenue DESC
LIMIT 5;
"""
result1 = pd.read_sql_query(query1, conn)
print("Q1: Top 5 Products by Total Revenue")
print("-" * 40)
print(result1)
```

```python
#2. what is the total profit  by category
query2="""
select "Category",round(sum(profit),2) as total_profit
from Superstore_Sales
group by "Category"
order by total_profit desc;
"""
result2=pd.read_sql_query(query2, conn)
print("Q2:Total profit  by category")
print("_"*40)
print(result2)
```

```python
# Q3. which Region has the highest total sales
query3="""
select "region",sum(sales) as total_sales
from Superstore_Sales
group by "region"
order by total_sales desc;
"""
result3=pd.read_sql_query(query3, conn)
print("Q3:Total sales by region")
print("_"*40)
print(result3)
```

```python
# Q4. which states are operating at a loss(Negative Profit)
query4="""
select "state",
sum(sales) as total_sales,
sum(profit) as total_profit
from superstore_sales
group by "state"
having total_profit<0
order by total_profit asc;
"""
result4=pd.read_sql_query(query4,conn)
print("Q4.states are operating at a loss(Negative Profit)")
print("_"*40)
print(result4)
```

```python
#Q5. How does Discount level Impact profit?
query5= """
select
  case
      when discount = 0 then 'No Discount'
      when discount <= 0.2 then 'low (0-20%)'
      when discount <=0.4 then 'medium (20-40)'
      else 'high (40%)'
      end as discount_level,
      round(avg(profit), 2) as avg_profit,
      round(avg(sales), 2) as avg_sales,
      count(*) as total_sales
  from superstore_sales
  group by discount_level
  order by avg_profit desc;
  """
result5=pd.read_sql_query(query5, conn)
print("Q5: Discount level Impact profit")
print("_"*40)
print(result5)
```

```python
# Q6. Which Product (Sub-Category) is Most Profitable?
query6 = """
select "sub-category",
    round(sum(sales), 2) as total_sales,
    round(sum(profit), 2) as total_profit,
    round((sum(profit) / sum(sales)) * 100, 2) as profit_margin_pct
from superstore_sales
group by "sub-category"
order by total_profit desc;
"""
result6 = pd.read_sql_query(query6, conn)
print("Q6. Which Product Sub-Category is Most Profitable?")
print("_" * 40)
print(result6)
```

```python
# Q7. Which Customer Segment Generates the Highest Revenue and Profit?
query7 = """
select segment,
    round(sum(sales), 2) as total_sales,
    round(sum(profit), 2) as total_profit,
    count(distinct "order id") as total_orders,
    round((sum(profit) / sum(sales)) * 100, 2) as profit_margin_pct
from superstore_sales
group by segment
order by total_profit desc;
"""
result7 = pd.read_sql_query(query7, conn)
print("Q7. Which Customer Segment Generates the Highest Revenue and Profit?")
print("_" * 40)
print(result7)
```

# STEP 3: Multivariate Analysis & Correlation

```python
# Import Required libraries #
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset #
df=pd.read_csv("Cleaned_dataset.csv")
df.shape
```

```python
df.columns
```

```python
# PLOT 1: Correlation Heatmap
#Pick only number columns #
num_cols=['Sales', 'Quantity','Discount', 'Profit']

# calculate correlation #
corr=df[num_cols].corr()
print(corr)

# Draw Heatmap
plt.figure(figsize=(7,5))
sns.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm",linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
# KEY FINDING: Discount has NEGATIVE correlation with profit(-0.22)
```

```python
# PLOT 2: Scatter Plot - Discount vs Profit
import plotly.express as px
# Create interactive scatter plot
fig = px.scatter(
    df,
    x="Discount",
    y="Profit",
    color="Category",
    title="Discount vs Profit",
    opacity=0.6,
    size_max=10
)

# Add horizontal line at Profit = 0
fig.add_hline(y=0, line_dash="dash", line_color="black")

# Show interactive chart
fig.show()
```

```python
# Plot 3: Scatter Plot - Sales vs Profit #

import plotly.express as px
fig=px.scatter(df,
                x='Sales',
                y='Profit',
                color="Category",
                title="Sales vs Profit",
                opacity=0.6,
                size_max=10)

fig.add_hline(y=0, line_dash='dash', line_color='black')

fig.show()

```

```python
# PLOT 4: Pair Plot
pair_df=df[['Sales', 'Quantity','Discount', 'Profit', 'Category']].copy()
pair_df=pair_df[pair_df['Sales']<5000]

g = sns.pairplot(
    pair_df,
    hue='Category',
    palette={'Furniture':'blue','Office Supplies':'green','Technology':'red'},
    plot_kws={'alpha': 0.3, 's': 15}
)

g.fig.suptitle('Pair Plot — All Variable Combinations', y=1.02, fontsize=13)
plt.show()
```
